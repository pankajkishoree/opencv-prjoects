```python
import cv2

# Load pretrained models for face and eye detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")

# Start capturing video
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # Capture a frame
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)  # Detect faces

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 255, 0), 2)  # Draw rectangle on face

        face_gray = gray[y: y + h, x: x + w]
        face_clr = frame[y: y + h, x: x + w]

        eyes = eye_cascade.detectMultiScale(face_gray)  # Detect eyes in face region
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(face_clr, (ex, ey), (ex + ew, ey + eh), (0, 127, 255), 2)  # Draw rectangle around eyes

    cv2.imshow("Face and Eye Detection", frame)

    if cv2.waitKey(1) == 27:  # Press 'Esc' key to exit
        break

# Release video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
